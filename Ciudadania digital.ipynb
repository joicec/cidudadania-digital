## Descripción de variables

Descripción de las variables indico que pregunta se le hizo a las personas para completar la variable.


ID: Identificador

Sexo: Sexo 

Departamento: ¿En qué departamento vive?

Nivel_educativo_Detallado:¿Cuál es el máximo nivel educativo cursado por usted?

Perdio_info: ¿Perdió información debido a un virus u otra infección informática?

Violaron_privacidad: ¿Violaron su privacidad por ejemplo, de fotos o videos o abusaron de información personal que suministró?

Hackearon: Le hackearon su correo electrónico o sus cuentas en redes sociales?, es decir, ¿alguien se hizo pasar por usted?

Ciber_Bullying: ¿Lo acosaron u hostigaron por internet por ejemplo ciber-bullying o chantaje?

Perdio_dinero: ¿Perdió dinero por fraude de tarjetas de crédito por Internet? No importa si recuperó o no el dinero

Dejo_de_comprar_por_inseguridad: ¿Dejó de comprar algún producto, servicio o contenido por internet por preocupaciones en temas de seguridad?

Cambio_password: En los últimos 3 meses ¿ha cambiado alguna de las contraseñas de correo electrónico, redes sociales, banca en línea o sitio de compras? 

Antivirus_compu: Esta computadora que usted utiliza principalmente, ¿Tiene antivirus? 

Pago_internet: ¿Alguna vez ha realizado pagos por internet? Hablamos únicamente de compras que usted pagó desde su computadora o celular, no las que luego tiene que ir a un Red pagos o Abitab 

Rango_edad: Rango de edad de los encuestados. (18 a 29, 30 a 59, 60 años y más)

Año: Año en que se realizo la encuesta (2021, 2022)

## CAMBIANDO DATOS PARA QUE SUS VALORES SEAN MAS DESCRIPTIVOS

```python
# no corresponde puede ser porque no tiene acceso a internet o no tiene computadora

df["Rango_edad"] = df.Rango_edad.replace({1: "18 a 29", 2: "30 a 59", 3: "60 y más"})

df['Nivel_educativo_Detallado'] = df.Nivel_educativo_Detallado.replace({1: "Primaria incompleta", 2: "Primaria completa", 3: "Ciclo Básico / UTU incompleta", 4: "Ciclo Básico / UTU completa", 5: "Bachillerato Incompleto", 6: "Bachillerato completo", 7: "Terciaria incompleta", 8: "Terciaria completa",  9: "Ns/Nc" }) 

df["Perdio_info"] = df.Perdio_info.replace({1: "Sí", 2: "No", 3: "Ns/Nc", -1: "No corresponde"})

df["Violaron_privacidad"] = df.Violaron_privacidad.replace({1: "Sí", 2: "No", 3: "Ns/Nc", -1: "No corresponde"})

df["Hackearon"] = df.Hackearon.replace({1: "Sí", 2: "No", 3: "Ns/Nc", -1: "No corresponde"})

df["Ciber_Bullying"] = df.Ciber_Bullying.replace({1: "Sí", 2: "No", 3: "Ns/Nc", -1: "No corresponde"})

df["Perdio_dinero"] = df.Perdio_dinero.replace({1: "Sí", 2: "No", 3: "Ns/Nc", -1: "No corresponde"})

df["Dejo_de_comprar_por_inseguridad"] = df.Dejo_de_comprar_por_inseguridad.replace({1: "Sí", 2: "No", 3: "Ns/Nc", -1: "No corresponde"})

df["Cambio_password"] = df.Cambio_password.replace({1: "Sí", 2: "No", 3: "Ns/Nc", -1: "No corresponde"})

df["Antivirus_compu"] = df.Antivirus_compu.replace({1: "Sí", 2: "No", 3: "Ns/Nc", -1: "No corresponde"})

df["Pago_internet"] = df.Pago_internet.replace({1: "Sí", 2: "No", 3: "Ns/Nc", -1: "No corresponde"})



```
